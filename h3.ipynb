```python
import pandas as pd
from utils import *

df = load_data()
```

```python
# User Acc at Step 3

for condition in range(1, 7):
    df_condition = df[df['condition'] == condition]
    data = {
        's1': [],
        # 's2': [],
        's3': [],
    }
    for index, row in df_condition.iterrows():
        s1,s2,s3=0,0,0
        for i in range(1, 21):
            if row[f"{i}_gt"] == row[f"{i}_1"]: s1 += 1
            # if row[f"{i}_gt"] == row[f"{i}_2"]: s2 += 1
            if row[f"{i}_gt"] == row[f"{i}_3"]: s3 += 1

        data['s1'].append(s1/20)
        # data['s2'].append(s2/20)
        data['s3'].append(s3/20)

    plot_unequal(data, f"User Acc btw s1 s2 - Condition {condition}")
```

```python
# User Acc at s2 and s3 given s1 was wrong

for condition in range(1, 7):
    df_condition = df[df['condition'] == condition]
    data = {
        's2': [],
        's3': [],
    }
    for index, row in df_condition.iterrows():
        s1,s2,s3=0,0,0
        for i in range(1, 21):
            if row[f"{i}_gt"] != row[f"{i}_1"]:
                if row[f"{i}_gt"] == row[f"{i}_2"]: s2 += 1
                if row[f"{i}_gt"] == row[f"{i}_3"]: s3 += 1

        data['s2'].append(s2/20)
        data['s3'].append(s3/20)

    plot_unequal(data, f"User Acc at s2/s3 given s1 was wrong - Condition {condition}")
```

```python
# User Acc at s3 given s1 and s2 were wrong

data = {
    'c1': [],
    'c2': [],
    'c3': [],
    'c4': [],
    'c5': [],
    'c6': [],
}
for condition in range(1, 7):
    df_condition = df[df['condition'] == condition]
    for index, row in df_condition.iterrows():
        s1 = 0
        s2 = 0
        for i in range(1, 21):
            if row[f"{i}_gt"] == row[f"{i}_1"]: s1 += 1
            if row[f"{i}_gt"] == row[f"{i}_2"]: s2 += 1
        data[f'c{condition}'].append(s2/20 - s1/20)

plot_unequal_multiple(data, f"User Acc change at s2 relative to s1")
```
