```python
df = load_data()
survey = ['satisfaction', 'complexity', 'reliability', 'trust', 'accurate']

def p_value_annotation(p_value):
    if p_value > 0.05:
        return "ns"
    elif p_value <= 0.05 and p_value > 0.01:
        return "*"
    elif p_value <= 0.01 and p_value > 0.001:
        return "**"
    elif p_value <= 0.001 and p_value > 0.0001:
        return "***"
    else:
        return "****"

def plot_pre_post_conditions(pre_scores, post_scores, question, condition, p_value):
    """
    Plots a box plot to compare pre- and post-scores for a specific question and condition.
    
    Parameters:
    pre_scores (numpy array): Pre-survey scores.
    post_scores (numpy array): Post-survey scores.
    question (str): The survey question being analyzed.
    condition (int): The condition number.
    p_value (float): P-value from the Wilcoxon signed-rank test.
    """
    fig, ax = plt.subplots(figsize=(6, 6))
    
    # Combine pre and post scores for boxplot
    data = [pre_scores, post_scores]
    labels = [f'Pre_{question}', f'Post_{question}']
    
    # Creating the box plot
    ax.boxplot(data, patch_artist=True, showmeans=True, widths=[0.5,0.5])
    
    # Adding titles and labels
    title = f'Pre-Post Comparison for Condition {condition}: {question}'
    ax.set_ylim(ymin=0, ymax=8)
    ax.set_title(title, fontsize=16)
    ax.set_xlabel(f'P = {p_value:.4f}', fontsize=14)
    ax.set_ylabel('Rating', fontsize=14)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    
    # Setting x-tick labels
    ax.set_xticklabels(labels)
    
    # Show grid for better readability
    # ax.grid(True, linestyle='--', alpha=0.7)
    
    y_max = max(np.max(pre_scores), np.max(post_scores)) + 0.3  # Add a small margin above max value
    ax.text(1.5, y_max, p_value_annotation(p_value), ha='center')
    ax.plot([1, 2], [y_max-0.05, y_max-0.05], color='black', linewidth=1.5)

    filename = "figures/" + title.replace(" ", "_").replace(":", "").replace("/", "") + ".png"
    plt.savefig(filename, bbox_inches='tight')
    
    # plt.show()

# Wilcoxon signed-rank test
for i in range(1, 7):
    temp = df[df['condition'] == i]
    print('Condition: ' + str(i))
    for q in survey:
        pre_scores = np.array([])
        post_scores = np.array([])
        for index, row in temp.iterrows():
            pre_scores = np.append(pre_scores, row['pre_' + q])
            post_scores = np.append(post_scores, row['post_' + q])
        # Get standard deviation of pre and post scores
        pre_mean = np.mean(pre_scores)
        post_mean = np.mean(post_scores)
        pre_std = np.std(pre_scores)
        post_std = np.std(post_scores)
        stat, p_value = wilcoxon(pre_scores, post_scores)
        sig = " *" if p_value < 0.05 else ""
        
        print(f"    pre_{q}: mean = {pre_mean:.5}, std = {pre_std:.5}")
        print(f"    post_{q}: mean = {post_mean:.5}, std = {post_std:.5}")
        print(f"    Wilcoxon signed-rank test: p-value = {p_value} {sig}")
        print()
        
        # Plot the pre-post comparison with significance
        plot_pre_post_conditions(pre_scores, post_scores, q, i, p_value)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
from scipy.stats import wilcoxon

def plot_pre_post_conditions(pre_scores, post_scores, question, condition, p_value, ax, show_ylabel=False):
    # Combine pre and post scores for boxplot
    data = [pre_scores, post_scores]
    labels = ['Pre', 'Post']
    
    # Calculate means and standard deviations
    pre_mean = np.mean(pre_scores)
    post_mean = np.mean(post_scores)
    pre_std = np.std(pre_scores)
    post_std = np.std(post_scores)
    
    # Set bar positions
    center = 1.0  # Center point of the x-axis
    gap = 0.5     # Gap between bars
    bar_width = 0.3
    bar_positions = [center - gap/2, center + gap/2]  # Evenly spaced around center
    
    # Create bars
    bars = ax.bar(bar_positions,
                 [pre_mean, post_mean],
                 bar_width,
                 yerr=[pre_std, post_std],
                 capsize=5,
                 color=['#D4F1F4', '#189AB4'],
                 error_kw={'ecolor': 'black', 'capthick': 1})
    
    # Set up the plot with extended y-axis for significance bars
    ax.set_ylim(ymin=0, ymax=8.5)  # Extended to 8.5 to give room for significance bars
    
    if condition == 1:  # Only show question titles for the first row
        ax.set_title(question.capitalize(), fontsize=12, pad=20)
    
    if show_ylabel:  # Only show y-label and condition number for leftmost plots
        # ax.set_ylabel(f'Condition {condition}', fontsize=12)
        ax.set_ylabel(f'{CONDITION_MAP[condition]}', fontsize=12)
    else:
        ax.set_ylabel('')
    
    # Remove top and right spines
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    
    # Setting x-tick labels
    margin = 0.5  # Margin around the bars
    ax.set_xlim(center - margin, center + margin)
    ax.set_xticks(bar_positions)
    ax.set_xticklabels(labels, fontsize=8)
    
    # Add significance annotation with more space
    # Position significance bar at 7.2 and text at 7.5
    sig_bar_height = 7.5
    sig_text_height = 7.7
    
    # Add the significance bar
    ax.plot([bar_positions[0], bar_positions[1]], 
            [sig_bar_height, sig_bar_height], 
            color='black', 
            linewidth=1.5)
    
    ax.plot([bar_positions[0], bar_positions[0]], 
            [sig_bar_height, sig_bar_height-0.2], 
            color='black', 
            linewidth=1.5)
    ax.plot([bar_positions[1], bar_positions[1]], 
            [sig_bar_height, sig_bar_height-0.2], 
            color='black', 
            linewidth=1.5)
    
    # Add the significance annotation
    ax.text(center, sig_text_height, 
            p_value_annotation(p_value), 
            ha='center',
            va='bottom')
    
    # Add major y-ticks only up to 7
    ax.set_yticks(range(0, 8))
    ax.set_yticklabels([str(i) if i <= 7 else '' for i in range(0, 8)])

def create_all_condition_plots(df, survey):
    # Set up the figure
    n_conditions = 6
    n_questions = len(survey)
    
    # Create figure with adjusted size
    fig = plt.figure(figsize=(3*n_questions, 3.5*n_conditions))  # Slightly taller to accommodate significance bars
    
    # Create grid of subplots with adjusted spacing
    gs = GridSpec(n_conditions, n_questions, figure=fig, 
                 hspace=0.4,  # Increased vertical spacing
                 wspace=0.4)
    
    # Plot each condition and question
    for condition in range(1, n_conditions + 1):
        temp = df[df['condition'] == condition]
        for q_idx, q in enumerate(survey):
            pre_scores = np.array([])
            post_scores = np.array([])
            
            for _, row in temp.iterrows():
                pre_scores = np.append(pre_scores, row['pre_' + q])
                post_scores = np.append(post_scores, row['post_' + q])
            
            # Calculate statistics
            stat, p_value = wilcoxon(pre_scores, post_scores)
            
            # Create subplot
            ax = fig.add_subplot(gs[condition-1, q_idx])
            
            # Only show y-label for leftmost plots
            show_ylabel = (q_idx == 0)
            
            plot_pre_post_conditions(
                pre_scores,
                post_scores,
                q,
                condition,
                p_value,
                ax,
                show_ylabel=show_ylabel
            )
    
    # Save the figure
    filename = "figures/All_Conditions_Combined_Plot.svg"
    plt.savefig(filename, bbox_inches='tight')
    plt.show()

# Create the combined plot
create_all_condition_plots(df, survey)
```

```python
import numpy as np
from scipy.stats import shapiro, kruskal, f_oneway, levene, mannwhitneyu

def kruskal_wallis_or_anova(conditions):
    # Step 1: Check normality for each condition
    normality_results = {}
    all_normal = True  # To track if all conditions are normally distributed
    for i, condition in enumerate(conditions, start=1):
        stat, p_value = shapiro(condition)
        normality_results[f'Condition {i}'] = p_value
        print(f'Condition {i} - Shapiro-Wilk p-value: {p_value}')
        if p_value < 0.05:
            all_normal = False

    # Step 2: Check for homogeneity of variance using Levene's test
    stat, p_value_levene = levene(*conditions)
    # print(f"Levene's test for equal variances p-value: {p_value_levene}")
    equal_variances = p_value_levene >= 0.05

    # Step 3: Decide whether to use ANOVA or Kruskal-Wallis
    if all_normal and equal_variances:
        # print("Data is normally distributed and variances are equal. Performing one-way ANOVA...")
        # Perform one-way ANOVA
        stat, p_value_anova = f_oneway(*conditions)
        p_value_anova_sig = " *" if p_value_anova < 0.05 else ""
        print(f"ANOVA test statistic: {stat} {p_value_anova_sig}")
        print(f"P-value: {p_value_anova}")
        
        if p_value_anova < 0.05:
            print("Significant result found in ANOVA. Performing post-hoc analysis with Tukey's HSD...")
            # Prepare data for Tukey's HSD
            data = np.concatenate(conditions)
            groups = np.array([f'Condition {i+1}' for i in range(len(conditions)) for _ in range(len(conditions[i]))])
            tukey_results = pairwise_tukeyhsd(data, groups)
            print("    " + tukey_results)
    else:
        # print("Normality or equal variance assumption violated. Performing Kruskal-Wallis test...")
        # Perform Kruskal-Wallis test
        stat, p_value_kruskal = kruskal(*conditions)
        p_value_kruskal_sig = " *" if p_value_kruskal < 0.05 else ""
        print(f"Kruskal-Wallis test statistic: {stat} {p_value_kruskal_sig}")
        print(f"P-value: {p_value_kruskal}")
        
        if p_value_kruskal < 0.05:
            print("Significant result found in Kruskal-Wallis test. Performing post-hoc pairwise Mann-Whitney U tests...")
            # Perform pairwise Mann-Whitney U tests
            num_conditions = len(conditions)
            for i in range(num_conditions):
                for j in range(i + 1, num_conditions):
                    stat, p_value_mannwhitney = mannwhitneyu(conditions[i], conditions[j])
                    p_value_mannwhitney_sig = " *" if p_value_mannwhitney < 0.05 else ""
                    print(f'    Condition {i+1} vs Condition {j+1} - Mann-Whitney U test p-value: {p_value_mannwhitney} {p_value_mannwhitney_sig}')
            print()

    # calculate mean and std for each condition
    for i, condition in enumerate(conditions, start=1):
        mean = np.mean(condition)
        std = np.std(condition)
        print(f"    Condition {i}: mean = {mean:.5}, std = {std:.5}")

questions = [
    "post_satisfaction",
    "post_complexity",
    "post_reliability",
    "post_accurate",
    "post_base_question",
    "post_condition_satisfaction",
    # post questionnaire 2
    "post_use",
    "post_confi",
    "post_understand_condition",
    "post_consistent",
    "post_demanding",
    "post_easier",
    "post_mental_demanding",
    "post_autonomy",
    "post_trust",
    "post_confident",
    "post_notice",
    "post_help_final",
    "post_increase_trust",
    "post_comfortable",
    "post_like_use",
    "post_trust_change"
]

for q in questions:
    print(f"Question: {q}:\n")
    conditions = []
    for c in range(1, 7):
        current = np.array([])
        temp = df[df['condition'] == c]
        for i, row in temp.iterrows():
            current = np.append(current, row[q])
        conditions.append(current)

    kruskal_wallis_or_anova(conditions)
    print()
```

```python
def kruskal_wallis_or_anova(conditions):
    significant_pairs = []  # List to store significant condition pairs

    # Step 1: Check normality for each condition
    normality_results = {}
    all_normal = True
    for i, condition in enumerate(conditions, start=1):
        stat, p_value = shapiro(condition)
        normality_results[f'Condition {i}'] = p_value
        if p_value < 0.05:
            all_normal = False

    # Step 2: Check for homogeneity of variance using Levene's test
    stat, p_value_levene = levene(*conditions)
    equal_variances = p_value_levene >= 0.05

    # Step 3: Decide whether to use ANOVA or Kruskal-Wallis
    if all_normal and equal_variances:
        stat, p_value_anova = f_oneway(*conditions)
        if p_value_anova < 0.05:
            # Perform post-hoc Tukey's HSD
            data = np.concatenate(conditions)
            groups = np.array([f'Condition {i+1}' for i in range(len(conditions)) for _ in range(len(conditions[i]))])
            tukey_results = pairwise_tukeyhsd(data, groups)

            # Collect significant pairs
            for comparison in tukey_results._results_table.data[1:]:
                group1, group2, p_value = comparison[0], comparison[1], comparison[5]
                if p_value < 0.05:
                    significant_pairs.append((group1, group2, p_value))
    else:
        stat, p_value_kruskal = kruskal(*conditions)
        if p_value_kruskal < 0.05:
            # Perform post-hoc pairwise Mann-Whitney U tests
            num_conditions = len(conditions)
            for i in range(num_conditions):
                for j in range(i + 1, num_conditions):
                    stat, p_value_mannwhitney = mannwhitneyu(conditions[i], conditions[j])
                    if p_value_mannwhitney < 0.05:
                        significant_pairs.append((f'Condition {i+1}', f'Condition {j+1}', p_value_mannwhitney))

    return significant_pairs

def plot_conditions(conditions, question_title, significant_pairs):
    # Set fixed figure size for consistency across all plots
    fig, ax = plt.subplots(figsize=(12, 8))
    
    # Number of conditions
    n_conditions = len(conditions)
    
    # Calculate means and standard errors for each condition
    means = [np.mean(condition) for condition in conditions]
    sems = [np.std(condition) / np.sqrt(len(condition)) for condition in conditions]
    
    # Define colors for each bar
    colors = ['#fd5901', '#f78104', '#faab36', '#249ea0', '#008083', '#005f60']
    while len(colors) < n_conditions:
        colors.extend(colors)
    colors = colors[:n_conditions]
    
    # Create bars
    x = np.arange(n_conditions)
    bars = ax.bar(x, means, yerr=sems, capsize=5, color=colors, alpha=0.7)
    
    # Calculate required y-axis limit based on number of significance pairs
    num_sig_pairs = len(significant_pairs)
    extra_space_needed = 0.25 * num_sig_pairs  # 0.25 units per significance bar
    y_limit = max(7, 7 + extra_space_needed)  # At least 7, or more if needed
    
    # Adding titles and labels with adjusted position
    ax.set_title(f'{question_title}', fontsize=16, pad=10)
    ax.set_ylabel('Rating', fontsize=14)
    
    # Remove top and right spines
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    
    # Setting custom x-tick labels
    ax.set_xticks(x)
    # ax.set_xticklabels([f'Condition {i+1}' for i in range(n_conditions)])
    # ax.set_xticklabels([f'{CONDITION_MAP[i+1]}' for i in range(n_conditions)])
    ax.set_xticklabels([f'{CONDITION_MAP[i+1]}' for i in range(n_conditions)], rotation=35, ha='right')
    
    # Set y-axis limits with dynamic upper limit
    ax.set_ylim(0, 8.5)  # Set limit to 8 to give space above

    # Set y-axis ticks to only show up to 7
    ax.set_yticks(np.arange(0, 7.1, 1))
    
    # Add significance markers
    y_start = 6.5  # Start position for first significance bar
    y_gap = 0.35   # Space between significance bars
    
    for i, pair in enumerate(significant_pairs):
        condition1 = int(pair[0].split()[-1]) - 1
        condition2 = int(pair[1].split()[-1]) - 1
        p_value = pair[2]
        
        # Plot significance bars
        x1, x2 = condition1, condition2
        y = y_start + (i * y_gap)  # Increment height for each pair
        
        # Draw the connecting line
        ax.plot([x1, x2], [y, y], color='black', linewidth=1)

        ax.plot([x1, x1], [y, y-0.1], color='black', linewidth=1)
        ax.plot([x2, x2], [y, y-0.1], color='black', linewidth=1)
        
        # Add asterisks based on p-value
        ax.text((x1 + x2) / 2, y, p_value_annotation(p_value), 
                ha='center', va='bottom')
    
    # Adjust layout to prevent cutting off significance markers and title
    plt.subplots_adjust(top=0.85, bottom=0.15, left=0.15, right=0.95)
    
    # Save figure with tight layout
    plt.savefig("figures/" + question_title.replace(' ', '') + ".svg", 
                bbox_inches='tight')
    
    return fig, ax


df = load_data()

questions = {
    "Confidence and Understanding": [
        "post_confi",
        "post_understand_condition" ],
    "Consistency and Cognitive Load": [
        "post_consistent",
        "post_demanding" ],
    "Ease and Usefulness": [
        "post_easier",
        "post_use",
        "post_like_use" ],
    "Trust and Comfort": [
        "post_trust",
        "post_increase_trust",
        "post_comfortable" ],
    "Trust Dynamics": [
        "post_trust_change"
    ]
}

for q in questions.keys():
    print(f"Question: {q}:\n")
    conditions = []
    for c in range(1, 7):
        current = np.array([])
        temp = df[df['condition'] == c]
        for i, row in temp.iterrows():
            temp = np.array([])
            for sub_q in questions[q]:
                temp = np.append(temp, row[sub_q])
            current = np.append(current, np.mean(temp))
        conditions.append(current)

    significant_pairs = kruskal_wallis_or_anova(conditions)
    print()
    plot_conditions(conditions, q, significant_pairs)
```
